## Image Manipulation

Manipulation of raw mobile phone captured images.

```python
import glob
import os

import matplotlib.pyplot as plt

from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing.image import load_img, img_to_array

from tensorflow.keras.utils import get_file
from PIL import Image
import time
from IPython.display import clear_output
```

### 1. Raw Image path.

```python
rawimages ='data/raw number plates/*.jpg'
```

```python
image_paths = list(glob.glob(rawimages))
```

```python
print(f'There are {len(image_paths):,} images in the dataset')
```

```python
for image in image_paths:
    img=Image.open(image)
    img = img.rotate(-90)

    width, height = img.size
    img = img.resize((width //5 ,height // 5), resample = Image.BICUBIC)

    display(img)
    print(img.size)
    time.sleep(5)
    clear_output(wait=False)
```

```python
output_image_path = 'data/resized/car_'
```

```python
for i,image in enumerate(image_paths):
    img=Image.open(image)
    img = img.rotate(-90)

    width, height = img.size
    img = img.resize((width //5 ,height // 5), resample = Image.BICUBIC)

    #display(img)
    //print(img.size)
    img.save(output_image_path+str(i)+'.jpg')
    #time.sleep(5)
    #clear_output(wait=False)
```
